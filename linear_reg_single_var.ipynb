```python
import pandas as pd
```

```python
df = pd.read_csv("home_prices.csv")
df.head()
```

```python
df[['area_sqr_ft']]
```

```python
df[['area_sqr_ft']].head()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(df[["area_sqr_ft"]], df['price_lakhs'])
```

```python
'''
as data size is smaller it taken few seconds.
usually in real world project data size will be big and takes some time
'''
```

```python
model.predict([[1500]])
```

```python
model.predict([[1500], [700]])
```

```python
model.coef_, model.intercept_
```

```python
model.coef_[0] * 1500 + model.intercept_
```
